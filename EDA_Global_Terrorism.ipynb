### change the columns name for ease to understanding

```python
data=data.rename(columns = {"iyear":"Year", "imonth":"Month", "iday":"Day","region_txt":"region_name","country_txt":"country_name",
                      "nwound":"wound", "targtype1_txt":"target_type","targtype1":"no_of_target_type","nkill":"kill"}, inplace = False)
```

```python
pd.set_option("display.max_columns", None)
print(data.dtypes)
```

### Extracting desire columns from dataset

```python
data=data[["Year", "Month", "Day", "country_name","region_name","country","city","wound", "gname", "kill",
           "target_type","no_of_target_type","weaptype1_txt","weaptype1","summary"]]
data.head()
```

### filling zero where null values in column

```python
data["kill"] = data["kill"].fillna(0)
data["wound"] = data["wound"].fillna(0)
data["summary"] = data["summary"].fillna(0)
```

### filling 'unknown' where columns have null values

```python
data["city"].fillna("Unknown", inplace = True)
```

```python
data.head(10)
```

```python
data.info()
```

### droping those columns whose have null values

```python
data.dropna(axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.info()
```

### Here are the 10 countries with the highest number of loses 

```python
plt.figure(figsize=(15,6))
plt.bar(list(data["country_name"].value_counts()[0:10].keys()), list(data["kill"].value_counts()[0:10]), color = "r")
plt.ylabel("No. Of Kills", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("Attacks on Country",fontsize = 18)
plt.title("Most kills Vs Country")
plt.show()
```

```python
kills_most = data[data["kill"] >= 1000]
```

```python
kills_most.head()
```

### Here are the 10 countries with the highest number of wounds 

```python
plt.figure(figsize=(15,6))
plt.bar(list(data["country_name"].value_counts()[0:10].keys()), list(data["wound"].value_counts()[0:10]), color = "y")
plt.ylabel("No. of Wounds", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("Attacks on Country",fontsize = 18)
plt.title("Most Wounds Vs Country")
plt.show()
```

```python
plt.rcParams["figure.figsize"] = 15,11
sns.barplot(x = data["country_name"].value_counts()[:10].index, y = data["country"].value_counts()[:10].values)
plt.ylabel("No. Of Attacks", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("Attacks on Country",fontsize = 18)
plt.title("NUmbers of Attacks On Country")
```

```python
plt.rcParams["figure.figsize"] = 15,11
sns.barplot(x = data["Year"].value_counts()[:10].index, y = data["country"].value_counts()[:10].values)
plt.ylabel("No. Of Attacks", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("Years",fontsize = 18)
```

```python

pd.crosstab(data.Year, data.region_name).plot(kind = "area", stacked = False, figsize = (20,10))
plt.title("Attacks on each year by region")
plt.ylabel("No. Of Attacks", fontsize = 18)
plt.xlabel("Year",fontsize = 18)
```

```python
plt.rcParams["figure.figsize"] = 15,11
sns.barplot(x = data["target_type"].value_counts()[:10].index, y = data["country"].value_counts()[:10].values)
plt.ylabel("No. Of Targets", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("Target By",fontsize = 18)
plt.title("Numbers of Attacks by Forces")
```

```python
plt.rcParams["figure.figsize"] = 15,11
sns.barplot(x = data["gname"].value_counts()[:10].index, y = data["gname"].value_counts()[:10].values)
plt.ylabel("No of Attacks", fontsize = 18)
plt.xticks(rotation = 90)
plt.xlabel("",fontsize = 18)
plt.title("Number Attaks by Terrorist Group")
```
